```python
import pandas as pd
import numpy as np

data = pd.read_excel('member_data.xlsx')
data
```

```python
len(data.아이디.unique())
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager

from selenium.webdriver.common.by import By

def set_chrome_driver():
    chrome_options = webdriver.ChromeOptions()
    driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()), options=chrome_options)
    return driver
```

```python
driver = set_chrome_driver()

for i in range(len(data)):
    driver.get("http://localhost:8899/bnk/register")
    
    driver.implicitly_wait(30)

    driver.find_element(By.ID, 'idForm').send_keys(data.아이디[i])
    driver.find_element(By.ID, 'pwForm').send_keys(data.비번[i])
    driver.find_element(By.NAME, 'memberName').send_keys(data.이름[i])
    driver.find_element(By.NAME, 'memberBirth').send_keys(''.join(['00', str(data.생년[i]), str(data.생월[i]), str(data.생일[i])]))
    if data.성별[i] == '남': driver.find_element(By.ID, 'male').click()
    else: driver.find_element(By.ID, 'female').click()
    driver.find_element(By.NAME, 'memberTele').send_keys(data.전번[i])
    driver.find_element(By.NAME, 'memberAddr').send_keys(data.주소[i])
    
    driver.find_element(By.ID, 'joinbutton').click()
    
    driver.implicitly_wait(30)
    print(str(data.이름[i]) + ' 완료')
```
